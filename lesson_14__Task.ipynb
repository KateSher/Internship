### Загрузка библиотек

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

### Загрузка данных из CSV-файла в датафрейм

```python
df = pd.read_csv('drive/MyDrive/Colab Notebooks/Internship/lesson_14/dataset.csv', names=['meanfreq',
                                                                                            'sd',
                                                                                            'median',
                                                                                            'Q25',
                                                                                            'Q75',
                                                                                            'IQR',
                                                                                            'skew',
                                                                                            'kurt',
                                                                                            'sp.ent',
                                                                                            'sfm',
                                                                                            'mode',
                                                                                            'centroid',
                                                                                            'meanfun',
                                                                                            'minfun',
                                                                                            'maxfun',
                                                                                            'meandom',
                                                                                            'mindom',
                                                                                            'maxdom',
                                                                                            'dfrange',
                                                                                            'modindx',
                                                                                            'label'], 
                                                                                            skiprows=1)
```

### Отображение первых 10 и последних 10 строк

```python
sliced_df = df.head(10)
sliced_df = sliced_df.append(df.tail(10))
sliced_df
```

### Разделение данных

```python
# координаты вектора (столбцы начиная с meanfreq и заканчивая modindx)
X=df.iloc[:, :-1]
print(X[5:8])
```

```python
# метка (label)
y=df.iloc[:,-1]
print(y[1580:1590])
```

### Декодирование данных y с помощью класса LabelEncoder

```python
y=df.iloc[:,-1]

# Encode label category
# male -> 1
# female -> 0

gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
print(y[1580:1590])
```

### Стандартизация данных X с помощью класса StandardScaler

```python
# Scale the data to be between -1 and 1

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
print(X[5:8])
```

### Разделение данных на обучающую (80%) и тестовую (20%) выборки

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

### Обучение алгоритма опорных векторов (kernel='linear'), вычисление метрики accuracy_score

```python
svc=SVC(kernel='linear') #Default hyperparameters
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

### Переобучение модели на новом ядре RBF (kernel='rbf'), вычисление метрики accuracy_score

```python
svc=SVC(kernel='rbf') #Default hyperparameters
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```
