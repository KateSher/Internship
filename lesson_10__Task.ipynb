```python
import scipy.integrate
from numpy import sin, cos, tan
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass
```

### Класс, выполняющий математическое описание криволинейных трапеций

```python
@dataclass
class Curvilinear_Trapezoid:
    a: float  # левая граница a
    b: float  # правая граница b
    f: str    # функция f(x)

    # построение графика функции f(x)
    def plot_func(self, a, b, f):
        x = np.linspace(a, b, 1000)
        y = list(map(lambda x: eval(f), x))
        plt.figure(figsize=(8, 6))
        plt.title('f(x) = ' + f, fontsize=20)
        plt.plot(x, y, linewidth=3, color = 'b')
        plt.vlines(a, 0, y[0], linewidth=3, color='r')
        plt.vlines(b, 0, y[-1], linewidth=3, color='g')
        plt.grid(True)
        plt.xlabel('x', fontsize=18)
        plt.ylabel('y', fontsize=18)
        plt.tick_params(which='both', labelsize=14)
        plt.show()

    # вычисление площади криволинейной трапеции
    def integrate(self, a, b, f):
        return scipy.integrate.quad(f, a, b)

# экземпляры класса для каждой функции f(x)
c1 = Curvilinear_Trapezoid(a=0, b=1.2, f='sin(x) + cos(2 * x)')
c2 = Curvilinear_Trapezoid(a=0.2, b=0.8, f='x * tan(x) + 2')
c3 = Curvilinear_Trapezoid(a=0, b=10, f='x**3 - 10 * x**2 + 170')
```

### Математическое описание криволинейных трапеций:

```python
print('Пример 1:', c1) 
print('Пример 2:', c2)
print('Пример 3:', c3)
```

### Вычисление площадей криволинейных трапеций. 
### Построение графиков функций f(x) в заданном диапазоне x ε [a; b]:

```python
# Пример 1: f(x) = sin(x) + cos(2 * x)
print('Площадь трапеции %f, точность расчетов %e' % \
      c1.integrate(a=0, b=1.2, f = lambda x: sin(x) + cos(2 * x)))

c1.plot_func(a=0, b=1.2, f='sin(x) + cos(2 * x)') 
```

```python
# Пример 2: f(x) = x * tan(x) + 2
print('Площадь трапеции %f, точность расчетов %e' % \
      c1.integrate(a=0.2, b=0.8, f = lambda x: x * tan(x) + 2))

c2.plot_func(a=0.2, b=0.8, f='x * tan(x) + 2') 
```

```python
# Пример 3: f(x) = x**3 - 10 * x**2 + 170
print('Площадь трапеции %f, точность расчетов %e' % \
      c1.integrate(a=0, b=10, f = lambda x: x**3 - 10 * x**2 + 170))

c3.plot_func(a=0, b=10, f='x**3 - 10 * x**2 + 170') 
```
