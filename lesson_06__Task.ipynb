```python
import matplotlib.pyplot as plt
import csv
import math
import matplotlib.ticker as ticker
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

### 1. Загрузка данных из файла dataset.txt

```python
x = []
y = []
 
with open('drive/MyDrive/Colab Notebooks/Internship/lesson_06/dataset.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=' ')
    
    next(plotting)
    for rows in plotting:
        x.append(int(rows[0]))
        y.append(float(rows[1]))
```

### 2. Построение графиков: линейный, ступенчатый, stem-график, столбчатая диаграмма

```python
# Линейный график

plt.figure(figsize=(15, 7))
plt.plot(x, y, label='Y = f(X)', color='teal', marker='o', markersize=4, markerfacecolor='greenyellow')
plt.title('Line graph', fontsize=22)

plt.xlabel('X', fontsize=16, color='maroon')
plt.ylabel('Y', rotation=0, fontsize=16, color='maroon')
plt.legend(fontsize=12)

plt.grid(True, which='major', linestyle='--', color='grey')
plt.minorticks_on()
plt.grid(True, which='minor', linestyle=':', color='grey', alpha=0.3)

plt.tick_params(which='both', width=2)
plt.tick_params(which='major', length=7, labelsize=12, labelcolor='darkviolet')
plt.tick_params(which='minor', length=4, color='darkslategrey')
plt.show()
```

```python
# Ступенчатый график 

fig, axs = plt.subplots(figsize=(22, 7))
axs.step(x, y, color='cornflowerblue', marker='o', where='post')
axs.set_facecolor('lightyellow')
axs.patch.set_alpha(0.2)
axs.set_title('Stair graph', fontsize=22)

axs.set_xlabel('X', fontsize=16)
axs.set_ylabel('Y', rotation=0, fontsize=16)

axs.grid(True, which='major', linestyle='--')
axs.minorticks_on()
axs.grid(True, which='minor', linestyle='--', alpha=0.4)

# установка интервала основных и вспомогательных делений:
axs.xaxis.set_major_locator(ticker.MultipleLocator(10))
axs.xaxis.set_minor_locator(ticker.MultipleLocator(1))

axs.tick_params(which='major', length=7, labelsize=12, labelcolor='mediumblue')
plt.show()
```

```python
# Stem-график

plt.figure(figsize=(15, 7))
plt.stem(x, y, use_line_collection=True, linefmt='g-', markerfmt='*')
plt.title('Stem graph', fontsize=22)

plt.xlabel('X', fontsize=16)
plt.ylabel('Y', rotation=0, fontsize=16)

plt.tick_params(which='major', length=7, labelsize=12)
plt.show()
```

```python
# Столбчатая диаграмма

plt.figure(figsize=(15, 7))
# RGB
color_rectangle = np.random.rand(100, 3)    
plt.bar(x, y, color = color_rectangle)
plt.title('Bar graph', fontsize=22)

plt.xlabel('X', fontsize=16)
plt.ylabel('Y', rotation=0, fontsize=16)

plt.tick_params(which='major', length=7, labelsize=12)
plt.show()
```

### 3. Нахождение максимального значения

```python
# Определение координат точки максимума

ymax = max(y)
xpos = y.index(ymax)
xmax = x[xpos]
print('max({}, {})'.format(xmax, ymax))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(22, 7))

# Построение графика Maximum value 

axs[0].set_title('Maximum value', fontsize=22)
axs[0].set_xlabel('X', fontsize=16, color='orangered')
axs[0].set_ylabel('Y', rotation=0, fontsize=16, color='orangered')

axs[0].grid(True, which='major', linestyle=':')
axs[0].tick_params(which='major', length=7, labelsize=12)

axs[0].annotate('max', xy=(xmax, ymax), xycoords='data',
              xytext=(30, 85),
              textcoords = 'data',
              fontsize=12, 
              arrowprops=dict(facecolor='orange'))
axs[0].plot(x, y, color='mediumslateblue')



# Построение графика Maximum value в увеличенном масштабе

axs[1].set_title('Maximum value (on an enlarged scale)', fontsize=22)
axs[1].set_xlabel('X', fontsize=16, color='orangered')
axs[1].set_ylabel('Y', rotation=0, fontsize=16, color='orangered')

axs[1].grid(True, which='major', linestyle=':')
axs[1].tick_params(which='major', length=7, labelsize=12)

axs[1].annotate('max \n({}, {})'.format(xmax, ymax), 
              xy=(xmax, ymax), 
              xycoords='data',
              xytext=(43, 93),
              ha='center',
              textcoords = 'data',
              fontsize=12,
              arrowprops=dict(facecolor='orange'))

axs[1].plot(x, y, color='mediumslateblue', linewidth=5, marker='o',
            markersize=5, markeredgewidth=3, markeredgecolor='deeppink')

axs[1].set_xlim([41, 48])
axs[1].set_ylim([86, 100])
plt.show()
```

### 4. Построение на одной подложке двух линейных графиков Y1 = f(X) и Y2 = 2*f(X)

```python
y1 = y
y2 = [i*2 for i in y]

fig, axs = plt.subplots(1, 2, figsize=(22, 7), facecolor='lavender')
fig.patch.set_alpha(0.4)

# Построение графика Y1 = f(X)

axs[0].plot(x, y1, label='Y1 = f(X)', linewidth=2, color='lightseagreen')
axs[0].set_facecolor('lavender')
axs[0].set_title('Y1 = f(X)', fontsize=22)

axs[0].set_xlabel('X', fontsize=16,)
axs[0].set_ylabel('Y1', fontsize=16,)

axs[0].tick_params(which='major', labelsize=12)
axs[0].grid(True, which='major', linewidth=2, color='w')
axs[0].legend(fontsize=12)



# Построение графика Y2 = 2*f(X)

axs[1].plot(x, y2, label='Y2 = 2*f(X)', linewidth=2, color='indigo')
axs[1].set_facecolor('lavender')
axs[1].set_title('Y2 = 2*f(X)', fontsize=22)

axs[1].set_xlabel('X', fontsize=16,)
axs[1].set_ylabel('Y2', fontsize=16,)

axs[1].tick_params(which='major', labelsize=12)
axs[1].grid(True, which='major', linewidth=2, color='w')
axs[1].legend(fontsize=12)
plt.show()
```
