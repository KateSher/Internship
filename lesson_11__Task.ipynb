# Загрузка данных из файла в виде NumPy-массива

```python
data = np.loadtxt('drive/MyDrive/Colab Notebooks/Internship/lesson_11/data.txt', dtype=str)
lst = []
for i in data:
    lst.append(float(i.replace(',', '.'))) # замена десятичного знака

arr=np.array(lst)
arr
```

# Рассчет для полученной выборки всех мер центральной тенденции, стандартного отклонения, дисперсии, размаха, максимального и минимального значения, кумулятивной суммы

```python
print('Медиана:', np.median(arr))
print('Средневзвешенное:', np.average(arr))
print('Среднее арифметическое:', np.mean(arr))
print('Стандартное отклонение:', np.std(arr))
print('Дисперсия:', np.var(arr))
print('Размах:', np.ptp(arr))
print('Максимальное значение:', np.amax(arr))
print('Минимальное значение:', np.amin(arr))
print('Кумулятивная сумма: \n', np.cumsum(arr))
```

# Построение гистограммы и функции распределения выборки

```python
plt.figure(figsize=(9, 10))

plt.hist(arr, bins=15, density=True, lw=1,
        ec='darkblue', fc='slateblue',
        alpha=0.6, label='Гистограмма выборки') 

plt.tick_params(labelsize=12)
plt.title('Гистограмма выборки', fontsize=22)
plt.legend(fontsize=14)

plt.xlabel('Данные', fontsize=18)
plt.ylabel('Частота', fontsize=18)
plt.grid(ls=':')
plt.show()
```

```python
plt.figure(figsize=(23, 10))
x = np.sort(arr)
y = np.arange(1, len(x) + 1) / len(x)
plt.plot(x, y, marker = '.', linestyle = 'none', label=r'F(x) = $\frac{n_{xi}}{n}$')

for i in np.arange(len(x)-1):
    plt.hlines(y[i], x[i], x[i+1])

plt.tick_params(labelsize=14)
plt.title('Функция распределения выборки', fontsize=22)
plt.legend(fontsize=24)

plt.xlabel('X', fontsize=18)
plt.ylabel('F(x)', rotation=0, fontsize=18)
plt.grid(ls=':')
plt.show()
```
